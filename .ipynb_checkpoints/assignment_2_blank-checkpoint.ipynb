## Define nodes and edges
some edges with an arrow indicate that they can only be traversed in one direction. 
For instance, one can traverse directly from Node 2 to Node 1 but one cannot traverse directly from Node 1 to
Node 2. Please consider taking a look at the assignment sheet for the position of the nodes and edges

```python
nodes = {
    '1': (4,1),
    '2': (2,2),
    '3': (4,4),
    '4': (6,3),
    '5': (5,6),
    '6': (7,5)
}
edges = {
    '1': [('3',5),('4',6)],
    '2': [('1',4),('3',4)],
    '3': [('1',5),('2',5),('5',6),('6',7)],
    '4': [('1',6),('3',5),('5',7)],
    '5': [('3',6),('4',8)],
    '6': [('3',7)]
}
```

```python
def draw_graph(nodes, graph):
    plt.figure()

    # Draw edges with weights (simple line + text)
    for node in graph:
        for neighbor, weight in graph[node]:
            x1, y1 = nodes[node]
            x2, y2 = nodes[neighbor]

            # line
            plt.plot([x1, x2], [y1, y2])

    # Draw nodes
    for node, (x, y) in nodes.items():
        plt.scatter(x, y, s=600)
        plt.text(x, y, node, ha='center', va='center')

    plt.title("Weighted Graph")
    plt.axis('equal')
    plt.show()
```

```python
draw_graph(nodes, edges)
```

## Problem formulation
This class defines a search problem for graph traversal.

- State: a node in the graph (e.g., '1', '2', '3')
- Action: moving from one node to its neighboring node
- Result: the next node reached after taking an action
- Goal test: checks whether the current node is the target node

This formulation allows BFS to operate on the graph using AIMA's search framework.

```python
graph = {
    '1': [('3'),('4')],
    '2': [('1'),('3')],
    '3': [('1'),('2'),('5'),('6')],
    '4': [('1'),('3'),('5')],
    '5': [('3'),('4')],
    '6': [('3')]
}
class GraphProblem(Problem):
    def __init__(self, initial, goal=None):
        super().__init__(initial, goal)

    def actions(self, state):
        return graph[state]

    def result(self, state, action):
        return action

    def goal_test(self, state):
        return state == self.goal
```

Heuristic Setup

```python
# NOTE: The heuristic (h) and path_cost methods below are needed for informed
# search (GBFS, AS, CUS2). If you move GraphProblem to the Problem Formulation
# section above, bring these methods with it.
class GraphProblem(Problem):
    """
    Extended GraphProblem that supports:
    - multiple destination goals (list)
    - actual edge costs via path_cost
    - Euclidean heuristic h(n) for informed search
    """
    def __init__(self, initial, goal=None):
        super().__init__(initial, goal)

    def actions(self, state):
        # Returns neighbour node IDs, sorted ascending for tie-breaking
        return sorted([n for n, _ in edges[state]])

    def result(self, state, action):
        return action

    def goal_test(self, state):
        # Supports both single goal (str) and multiple goals (list)
        if isinstance(self.goal, list):
            return state in self.goal
        return state == self.goal

    def path_cost(self, c, state1, action, state2):
        # Look up actual edge weight between state1 and state2
        for neighbour, cost in edges[state1]:
            if neighbour == state2:
                return c + cost
        return c + math.inf

    def h(self, node):
        # Heuristic: min Euclidean distance from current node to any goal
        x1, y1 = nodes[node.state]
        goals = self.goal if isinstance(self.goal, list) else [self.goal]
        return min(
            math.sqrt((x1 - nodes[d][0])**2 + (y1 - nodes[d][1])**2)
            for d in goals
        )
```

Helpers

```python
# ─── HELPER: extract path and format output ───────────────────────────────────
def get_path(solution_node):
    """Return list of node IDs from root to goal."""
    if solution_node is None:
        return None
    return [n.state for n in solution_node.path()]

def format_path(path):
    return ' -> '.join(path) if path else 'No solution'

# ─── HELPER: draw graph with path highlighted ─────────────────────────────────
# NOTE: This is an enhanced version of draw_graph from above.
# The original draw_graph is kept unchanged; this one is only used
# in the solve section to highlight search results.
def draw_solution(nodes, graph, path=None, title='Result'):
    path_edges = set(zip(path[:-1], path[1:])) if path else set()
    plt.figure()
    for node in graph:
        for neighbor, weight in graph[node]:
            x1, y1 = nodes[node]
            x2, y2 = nodes[neighbor]
            color = 'blue' if (node, neighbor) in path_edges else 'lightgray'
            lw = 2.5 if (node, neighbor) in path_edges else 1.0
            plt.annotate('', xy=(x2, y2), xytext=(x1, y1),
                         arrowprops=dict(arrowstyle='->', color=color, lw=lw))
            mx, my = (x1+x2)/2, (y1+y2)/2
            plt.text(mx, my, str(weight), fontsize=8, color='gray', ha='center', va='center')
    for node, (x, y) in nodes.items():
        color = 'lightblue' if path and node in path else 'white'
        plt.scatter(x, y, s=600, c=color, edgecolors='black', zorder=3)
        plt.text(x, y, node, ha='center', va='center', fontweight='bold', zorder=4)
    plt.title(title)
    plt.axis('equal')
    plt.show()
```

Breadth First Search (BFS)

```python
problem = GraphProblem('2', '5')
solution = breadth_first_graph_search(problem)
print('BFS')
print('Actions:', solution.solution())
print('Path:   ', format_path(get_path(solution)))
print('Cost:   ', solution.path_cost)
draw_solution(nodes, edges, get_path(solution), title='BFS — Breadth-First Search')
```

Depth First Search (DFS)

```python
problem = GraphProblem('2', '5')
solution = depth_first_graph_search(problem)
print('DFS')
print('Actions:', solution.solution())
print('Path:   ', format_path(get_path(solution)))
print('Cost:   ', solution.path_cost)
draw_solution(nodes, edges, get_path(solution), title='DFS — Depth-First Search')
```

Greedy Best First Search (GBFS)

```python
problem = GraphProblem('2', '5')
solution = greedy_best_first_graph_search(problem)
print('GBFS')
print('Actions:', solution.solution())
print('Path:   ', format_path(get_path(solution)))
print('Cost:   ', solution.path_cost)
draw_solution(nodes, edges, get_path(solution), title='GBFS — Greedy Best-First Search')
```

A* Search (AS)

```python
problem = GraphProblem('2', '5')
solution = astar_search(problem)
print('A*')
print('Actions:', solution.solution())
print('Path:   ', format_path(get_path(solution)))
print('Cost:   ', solution.path_cost)
draw_solution(nodes, edges, get_path(solution), title='AS — A* Search')
```

Iterative Deepening DFS (uninformed) (CUS1)

```python
problem = GraphProblem('2', '5')
solution = iterative_deepening_search(problem)
print('CUS1 — Iterative Deepening DFS')
print('Actions:', solution.solution())
print('Path:   ', format_path(get_path(solution)))
print('Cost:   ', solution.path_cost)
draw_solution(nodes, edges, get_path(solution), title='CUS1 — Iterative Deepening DFS')
```

Weighted A* (informed) (CUS2)

```python
# weight=1.5 means the heuristic is scaled up, making search faster but
# potentially slightly less optimal than standard A* (weight=1.0)
problem = GraphProblem('2', '5')
solution = weighted_astar_search(problem, weight=1.5)
print('CUS2 — Weighted A* (w=1.5)')
print('Actions:', solution.solution())
print('Path:   ', format_path(get_path(solution)))
print('Cost:   ', solution.path_cost)
draw_solution(nodes, edges, get_path(solution), title='CUS2 — Weighted A* (w=1.5)')
```
